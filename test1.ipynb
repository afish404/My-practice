```python
def use_sorted():
    L = [4,1,3,2]
    sorted_L = sorted(L)
    print(L)
    print(sorted_L)
    S = "python"
    print(sorted(S))

def length(element):
    return len(element)
    

def str_sorted():
    words = ["apple", "banana", "cat", "dog"]
    sorted_words = sorted(words,key=length)
    print(sorted_words)

def func1(x):
    return x['age']

def dict_sorted():
    students = [
        {"name": "Alice", "age": 18},
        {"name": "Bob", "age": 16},
        {"name": "Charlie", "age": 20}
    ]
    print(sorted(students,key=lambda x:x['age']))

def two_column():
    tup = [(3, 5), (1, 2), (2, 4), (3, 1), (1, 3)]
    print(sorted(tup,key=lambda x:(x[0],-x[1])))
    students = [
        {"name": "Alice", "age": 18},
        {"name": "Bob", "age": 18},
        {"name": "Charlie", "age": 20}
    ]
    print(sorted(students,lambda x:(x['age'],len(x['name']))))

if __name__ == "__main__":
    use_sorted()
    str_sorted()
    dict_sorted()
    two_column()
```
