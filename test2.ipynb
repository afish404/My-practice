```python
import random
from matplotlib import  font_manager
font_manager.fontManager.addfont('C:/Windows/Fonts/simhei.ttf')

x = range(2,26,2)
y = [random.randint(20,30) for i in x]
plt.figure(figsize=(20,15),dpi=80)

x_ticks = range(2,26,2)
x_ticks_labels = ['{}:00'.format(i) for i in x]
plt.xticks(x_ticks,x_ticks_labels,fontsize=10,rotation=45)
y_ticks = range(min(y),max(y)+1)
y_ticks_labels = ['{}'.format(i) for i in y_ticks]
plt.yticks(y_ticks,y_ticks_labels)

my_font = font_manager.FontProperties(fname='C:/Windows/Fonts/simhei.ttf',size=20)
plt.title('',fontproperties=my_font)
plt.xlabel('',fontproperties=my_font)
plt.ylabel('',fontproperties=my_font)
plt.plot(x,y)
plt.savefig('test.svg')
```
