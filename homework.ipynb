```python
#4、完成apply和map的练习

data = {
    "name": ["Alice", "Bob", "Charlie", "David"],
    "age": [20, 22, 21, 20],
    "score": [85, 90, 78, 92],
    "gender": ["女", "男", "男", "男"]
}
df = pd.DataFrame(data, index=['s1', 's2', 's3', 's4'], columns=['name', 'age', 'score', 'gender'])

def grow_up(age):
    return age + 1

df['age'] = df['age'].apply(grow_up)
print(df)

df = pd.DataFrame(np.arange(12).reshape(3,4), index=['s1', 's2', 's3'], columns=['c1', 'c2', 'c3', 'c4'])
print(df.apply(lambda x: x.count(),axis=0))
print(df.apply(lambda x: x.max(),axis=0))
print(df.apply(lambda x: x.count(),axis=1))
```

```python
#4、完成apply和map的练习

data = {
    "name": ["Alice", "Bob", "Charlie", "David"],
    "age": [20, 22, 21, 20],
    "score": [85, 90, 78, 92],
    "gender": ["女", "男", "男", "男"]
}
df = pd.DataFrame(data, index=['s1', 's2', 's3', 's4'], columns=['name', 'age', 'score', 'gender'])

gender_excert = {
    "男": "male",
    "女": "female"
}
df['gender'] = df['gender'].map(gender_excert)
print(df)
df['gender'] = df["gender"].map(lambda x: 1 if x == 'male' else 0)
print(df)


df2 = pd.DataFrame(np.arange(12).reshape(3,4), index=['s1', 's2', 's3'], columns=['c1', 'c2', 'c3', 'c4'])
print(df2.map(lambda x: x*2))


df_float = pd.DataFrame(np.random.randn(3, 4), columns=['c1', 'c2', 'c3', 'c4'])
def my_round(x):
    return f'{x:0.2f}'
df_float2 = df_float.map(my_round)
print(df_float2)
```

6、编写机器学习的训练过程，什么是特征，什么是标签，什么是训练集，什么是测试集，
什么是有监督，无监督学习，什么是分类问题，回归问题，什么是欠拟合和过拟合，
以问答题写到代码文件里即可。

特征：特征是各组数据中，具有实际意义能够代表某一具体属性的一些列（一般为一列）值或字符
标签：给数据提供的值
训练集：用于训练的一组数据
测试集：用于检测训练结果的一组数据
有监督：所输入的数据是具有标签的的，一般课分为分类和回归
无监督：所输入的数据是不具有标签的
分类问题：预测结果为一个类别，所呈现结果为离散值
回归问题：预测结果为一个连续值
欠拟合：测试集和训练集所呈现效果均不佳
过拟合：测试集效果好但训练集效果不行
